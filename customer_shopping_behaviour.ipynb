```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a new column age_group
labels=['young_adult','adult','middle_aged','senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#purchase frequency rate column
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':356,
    'Every 3 Months':90}
df['purchase_frequency_rate']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# check is disount is done when promocode used
df[['purchase_frequency_rate','frequency_of_purchases']].head(10)
```

```python
#yes discount is always applied when promocode used  wwe dont need this column 

df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```
